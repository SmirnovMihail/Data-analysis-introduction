## 19% - опоздавшие рейсы в тренировочной выборке

```python
def distribution(feat_name):
    feat = data_train[data_train['dep_delayed_15min'] == 1][feat_name].value_counts()
    table = pd.DataFrame(columns=[feat_name, 'delayed', 'total', '%'])
    
    for j, i in enumerate(feat.index):
        a = feat[i]
        b = data_train[data_train[feat_name] == i][feat_name].count()
        c = round((a/b)*100, 1)
        
        table.loc[j] = [i,a,b,c]
        
    table.sort_values(by='%',ascending=False ,inplace=True)
    return table
```

## Но сначала onehotencoding на всех признаках

```python
def dummies_converter(X, column):
    dummy_feature = pd.get_dummies(X[column], prefix=column)
    X = pd.concat([X, dummy_feature], axis=1)
    X = X.drop([column], axis=1)
    return X
```

```python
data_train
```

```python
y = data_train['dep_delayed_15min']
data_train.drop(['dep_delayed_15min', 'DepTime'], axis = 1, inplace=True)
```

```python
obj_feat.append('Month')
obj_feat.append('DayofMonth')
obj_feat.append('DayOfWeek')
obj_feat.append('hour')
```
